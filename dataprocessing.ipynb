```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
states = gpd.read_file('data/gz_2010_us_040_00_500k.json',geometry='geometry')
states.info()
```

```python
wnv_us = pd.read_csv('data/Book1.csv')
wnv_us.info()
```

```python
df = states.merge(wnv_us, left_on='NAME',right_on="State")
df = df[['State', 'Total cases', 'Deaths', 'geometry']]
df.head()
```

```python
df["Total cases"]=df["Total cases"].astype(np.int)
df["Deaths"] =  df["Deaths"].astype(np.int)
```

```python
df.to_file('data/wnv_states.geojson')
```
